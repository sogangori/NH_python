```python
import csv #  모듈 호출
f = open('seoul.csv', 'r', encoding='cp949') #파일 여는 내장 함수
data = csv.reader(f, delimiter=',') # , 구분자 
type(data)
```

```python
#p 39 헤더와 데이터를 분리
header = next(data)
header
```

```python
# p 37  누락데이터 예 ['1952-07-12', '108', '', '', '']
# 목표 : 최고 온도였던 날짜가 언제인가를 찾아보세요 
max_v = -100
max_data = ''
for row in data:
    max_temp = row[-1]
    if max_temp != '': #누락 데이터가 아니면
        if float(max_temp) > max_v : #지금까지의 최고 기온보다 높으면
            max_v = float(max_temp)
            max_data = row[0]
print(max_data, 'max_v', max_v)
```
