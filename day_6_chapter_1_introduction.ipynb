```python
#p44
from sklearn.model_selection import train_test_split #함수
```

```python
x = iris_dataset.data
y = iris_dataset.target
```

```python
# 데이터를 학습용 75% / 테스트용 25% 로 나눠서 섞습니다, 
# 실험과 증명과 재현을 위해서 random_state 를 사용
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0)
x.shape , y.shape
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
y_test
```

```python
#45 산점도 scatter #46
df = pd.DataFrame(x_train, columns=iris_dataset.feature_names)
df.head()
```

```python
plt.rc('font', family='Malgun Gothic')
p = df.plot()
```

```python
#47 k-최근접 이웃 모델 
from sklearn.neighbors import KNeighborsClassifier #대문자로 시작하면 클래스
```

```python
knn = KNeighborsClassifier(n_neighbors=1) #분류기 객체 생성
```

```python
#48 훈련 데이터셋으로 모델을 만드는 .fit 함수
knn.fit(x_train, y_train) # fit(모델을 훈련 데이터에 맞게 핏팅시켜라)
```

```python
p = plt.title(iris_dataset.feature_names[0])
p = plt.scatter(np.arange(len(x_train)), x_train[:,0], c=y_train, cmap=mglearn.cm3)
```

```python
#p48 예측하기
x_new = np.array([[5, 2.9, 1, 0.2]]) # 야생에서 꽃을 발견했다!
x_new.shape
```

```python
#p49
prediction = knn.predict(x_new) #클래스를 예측(분류)해라
print('예측', prediction)
print('예측한 타겟의 이름은 ', iris_dataset.target_names[prediction])
```

```python
#49 모델 평가하기 
y_pred = knn.predict(x_test)
print("테스트 세트에 대한 예측값 ")
print('예측', y_pred)
print('정답', y_test)
```

```python
print('정확도', knn.score(x_test, y_test) ) 
```
