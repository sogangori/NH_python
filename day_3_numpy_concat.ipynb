```python
import pandas as pd
import numpy as np
```

```python
df0 = pd.read_csv('^DJI.csv')
df1 = pd.read_csv('^DJI (1).csv')
df0.shape, df1.shape
```

```python
arr0 = np.array(df0)
arr1 = np.array(df1)
```

```python
arr = np.concatenate((arr1, arr0), axis=0)
arr.shape
```

```python
arr[0]
```

- 각 컬럼의 평균값을 구합니다
- 가격 변동(Close-Open)이 가장 큰 때가 언제인지(날짜)를 알고 싶다.

```python
np.mean(arr[:, 1:], axis=0).astype(np.int) #날짜 빼주세요
```

```python
Close = arr[:,4]
Open =  arr[:,1]
Gap = np.abs(Close - Open) #절대값을 구하세요. 아니면 가격이 가장 많이 상승한 날이 조회됩니다
index = np.argmax(Gap) #변동이 가장 큰 행번호
np.max(Gap), (Close - Open)[index], arr[index, 0] #변동이 가장 큰 행의 날짜 정보
```

중국 미국 무역전쟁으로 2018년 2월 2일부터 전세계 주가가 20%이상 폭락했고 다우지수는 133년 중 최대 낙폭 기록
